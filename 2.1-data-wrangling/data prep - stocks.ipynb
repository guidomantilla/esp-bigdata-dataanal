```python
import pandas as pd
import numpy as np
```

```python
profiles = pd.read_parquet("profiles.parquet")
monthly_prices = pd.read_parquet("monthly_prices.parquet")
```

```python
del monthly_prices["return"]
```

```python
monthly_prices.info()
```

```python
profiles.info()
```

```python
data = (profiles.loc[(profiles.isUS == True) & (profiles.isNAPrimary == True), ["symbol", "companyName"]]
        .merge(monthly_prices, on="symbol")
       )
```

```python
data = data.loc[data.date>="2017-01-01"].reset_index(drop=True)
```

```python
data.to_parquet("monthly_stocks.parquet", index=False)
```
