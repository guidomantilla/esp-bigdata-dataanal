#### Alias/Referencias en las listas
- Son mutables y las asignaciones a un objeto alteran el primero
- Pasar listas a funciones puede suponer un riesgo

````python
lista = [2, 4, 16, 32]
ref = lista

ref[2] = 64

print(ref)
print(lista)
````

````python
lista = [2, 4, 16, 32]
copia = lista[:]
# copia = lista.copy()

copia[2] = 64

print(lista)
print(copia)
````

````python
# listas anidadas se copian por referencia
lista = [2, 4, 16, 32, [34, 10, [5, 5]]]
copia = lista.copy()

copia[4][0] = 28

print(copia)
print(lista)
````

````python
# evitar que listas anidadas se copien por referencia
import copy

lista = [2, 4, 16, 32, [34, 10, [5, 5]]]

copia = copy.deepcopy(lista)
copia[0] = 454
copia[4][2][0] = 64
print(lista)
print(copia)
````

#### Bucle 'while'

* Repetici√≥n de un bloque de c√≥digo hasta que se deje cumplir una expresi√≥n (es decir, hasta que una condici√≥n evalue a *False*).
* Si la condici√≥n evalua a *False* desde el principio, el bloque de c√≥digo nunca se ejecuta.
* Cuidado con los bucles infinitos.

Formato general:

```
while test:       # Mientras se cumple la condici√≥n
    statements    # Instrucciones a ejecutar
```

````python
# Ejemplo: Mostrar los primeros 3 objetos de una lista

indice = 0
numeros = [9, 4, 7, 1, 2]

while indice < 3:
    print(numeros[indice])
    indice += 1
````

````python
# Ejemplo: contar y mostrar los n√∫meros inferiores a 10.

numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0

while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1

print("Contador:", contador)
print("Indice:", indice)
````

## Functions

## ¬øQu√© son?

* Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera que se puedan ejecutar cuando el programador considere oportuno.
* Las funciones van identificadas por un nombre.
* Se debe usar este nombre para invocar (ejecutar) el c√≥digo de la funci√≥n.
* Las funciones calculan un resultado (output) en base a unos par√°metros de entrada (input).
* En cada invocaci√≥n de la funci√≥n, los par√°metros de entrada pueden variar.

````python
def sumar(x, y):
    return x + y


print(sumar(4, 5))
print(sumar(-2, 0))
````

````python
# f = c √ó 1.793 + 32

c = 30
print(c * 1.793 + 32)  # celsius to fahren

c = 26
print(c * 1.793 + 32)
````

````python
def convert_celsius_to_fahrenheit(c):
    return c * 1.791117 + 32
````


## ¬øQu√© ventajas nos proporcionan?

* Eliminaci√≥n de duplicaci√≥n de c√≥digo.
* Incremento de la reutilizaci√≥n de c√≥digo.
* Manejo de complejidad: permiten descomponer grandes sistemas en piezas m√°s peque√±as y, por tanto, m√°s manejables y comprensibles.

Las funciones se definen por medio de la palabra clave *def*.

Formato general:

```
def name(arg1, arg2, ..., argN):
    statements
```

Muy com√∫nmente:

```
def name(arg1, arg2, ..., argN):
    ...
    return value
```

````python
def div_safe(x, y):
    if y == 0:
        return 0
    return x // y


print(div_safe(6, 2))
print(div_safe(6, 0))
````

````python
def div_safe(x, y):
    if y != 0:
        return x // y
    else:
        return 0


print(div_safe(6, 2))
````

````python
import numpy as np
import matplotlib.pyplot as plt


def f(datos, color, d):
    """
    Funci√≥n que calcula estad√≠sticas b√°sicas y grafica un histograma.

    :param datos: Lista o array de datos num√©ricos.
    :param color: Color del histograma.
    :param d: Cantidad de decimales en los resultados.
    :return: Lista con [mediana, promedio, m√≠nimo, m√°ximo]
    """
    # 1. Calcular valores estad√≠sticos
    mediana = round(np.median(datos), d)
    promedio = round(np.mean(datos), d)
    minimo = round(np.min(datos), d)
    maximo = round(np.max(datos), d)

    # 2. Graficar histograma
    plt.figure()  # Crea una nueva ventana de figura
    plt.hist(
        datos, color=color, edgecolor="black", alpha=0.7
    )  # Histograma con bordes negros y transparencia
    plt.title("Histograma de Datos")
    plt.xlabel("Valores")
    plt.ylabel("Frecuencia")
    plt.grid(True, linestyle="--", alpha=0.5)  # Cuadr√≠cula tenue
    plt.show()

    # 3. Retornar resultados
    return [mediana, promedio, minimo, maximo]


# Ejemplo de uso
x = np.random.randn(100)  # Simulaci√≥n de datos aleatorios
resultado = f(x, "blue", 3)
print("Mediana, Promedio, M√≠nimo, M√°ximo:", resultado)
````
