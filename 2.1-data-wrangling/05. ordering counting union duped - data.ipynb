## Find duplicates and removing them

* observaciones iid
* sobre ajuste en el modelo pues el peso de los datos duplicados generan patrones erroneos.


```python
mask1 = data.duplicated(subset = ["Name", "Street Address", "City"] )
```

```python
mask1
```

```python
data.loc[mask1]
```

```python
data.loc[mask1].loc[data.Name == "Tierra Mia Coffee"]
```

```python
mask = data.duplicated(subset = ["Name", "Street Address", "City"], keep = 'first' ) #keep ='last' # keep : Identifica la ocurrencia de duplicados despues del primero
# df = df[~df.index.duplicated(keep='first')] ## Genera un indice boolenao, que identifica en que momento ocurren un duplicado entre los index.
```

```python
data.loc[mask]
```

```python
data.loc[mask].loc[data.Name == "Tierra Mia Coffee"]
```

```python
mask.head()
```

```python
data_clean = data.loc[~mask] ## Filtra el df excluyendo las filas donde mask = True.
```

```python
data_clean.head()
```

```python
data_clean.shape
```

```python
data_clean.loc[data_clean.Name == 'Fire Wings']
```

```python
## Elimminación de los registros duplicados
data_clean2 = data.drop_duplicates(subset = ["Name", "Street Address", "City"], keep = 'last' ) #last
## data.drop_duplicated(keep = 'first') ## recorre todas las columnas del data frame
data_clean2.loc[data_clean2.Name == 'Fire Wings']
```

```python
data_clean2.shape
```

```python
data_clean2["Name"].value_counts()
```

```python
## Revisemos que pasa con los count mayores a 1; pueden compartir el mismo nombre pero la ciudad puede variar o el la dirección dado el caso
data_clean2[data_clean2.Name == "Omaniku"]
```
