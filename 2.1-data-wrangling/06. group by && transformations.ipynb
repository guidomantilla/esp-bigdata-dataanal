```python
data = data.sort_values(["symbol", "date"])
data
```

```python
data["monthly_return"]=data.groupby("symbol")["close"].pct_change(1)
```

```python
data.groupby("symbol")["close"].transform("pct_change")
```

```python
# transform on the entire DF
data["close"].transform(["pct_change", "cumsum"])
```

```python
# Transform in Groupby is less powerful:
data.groupby("symbol")["close"].transform("pct_change")
```

```python
data.groupby("symbol")["close"].mean()
```

```python
data["temp"] = data.monthly_return + 1
data
```

```python
data["cumprod"] = data.groupby("symbol")["temp"].cumprod()
```

```python
data["cummulative_return"] = data["cumprod"] - 1
data
```
