### Selector .loc

* .loc y .iloc son metodos utilizados para acceder y manipular datos en un marco de datos

* .loc selecciona datos basandose en las etiquetas del index y el nombre de las columnas.
* .iloc Selecciona datos basado en la posición numerica de las filas y columnas

```python
data.loc[10]
```

```python
### Selector .loc (significa ubicación)
data.loc[0:10,"Date":"Total Volume"]
```

```python
data.loc[[0, 10],["Date","Total Volume"]]
```
