# Pandas Groupby

```python
import pandas as pd
import numpy as np
```

```python
realtor_data = pd.read_csv("realtor-data.csv", parse_dates=["prev_sold_date"]) ## Base de agentes inmobiliarios
restaurants = pd.read_csv("/content/california_restaurants.csv")
```

```python
realtor_data.sample(2)
```

* ¿Que ciudades tienen las casas más caras y más baratas en EEUU?
* ¿Que combinacione de dormitorio y baño son las más comunes ?

```python
## Agrupemos los datos, esto nos indica que es una data frame agrupado
grp = realtor_data.groupby("city")
grp
```

```python
## Para conocer los grupos debemos iterar a través de los grupo y observar la respuesta real
for city, data in grp:
  print(city)
  print(data.shape[0])
```

```python
## Ahora, para volver más claro el ejericico agrupemos ciudad con estado

realtor_data["city_state"] = realtor_data["city"] + "," + realtor_data["state"]
grp = realtor_data.groupby("city_state", as_index= False) ## as_index = False indica que la variable agrupacion ciudad-estado no la asigne como indice
```

```python
realtor_data.sample(3)
```

```python
for city, data in grp:
  print(city)
  print(data.shape[0])
```

```python
## Que ciudades tienen las casas caras y baratas

median_price_city = grp["price"].median()

median_price_city
```

```python
min_max_city = grp["price"].agg(["median", "mean", "max", "min", "std", "count"])
min_max_city
```

```python
min_max_city.sort_values("median", ascending=False, inplace=True)
min_max_city
```

```python
## veamos ahora cantidad de valores por estados
realtor_data["state"].value_counts()
```

```python
## Utilicemos consultas anidadas o varias al tiempo
home_sizes_by_city= (grp["house_size"]
    .agg(["median", "mean", "max", "min", "std"])
    .sort_values("median", ascending=False)
)
home_sizes_by_city
### En la interpretación las primeras corresponden a las casas más grandes.
```

```python
## Ahora respondamos a la pregunta ¿Qué combianción de dormitorios y baños son los más comunes en data frame ?, ej 1 habitacion y 2 baños
realtor_data.groupby(["bed", "bath"], as_index=False)["city"].count() ## realiza el conteo por ciudad
```

```python
## Ahora la consulta anterior no responde lo que necesitamos
counts = realtor_data.groupby(["bed", "bath"], as_index=False)["city"].count().rename({"city":"count"}, axis = 1)
```

```python
counts.sort_values("count", ascending=False)

## si es nuestro interes organizar los index dados que están desordenados (reset_index)
counts = counts.sort_values("count", ascending=False).reset_index(drop=True)
counts
```

```python
## Como reorganizamos el codigo anterior
counts = (
    realtor_data.groupby(["bed", "bath"], as_index=False)["city"]
    .count()
    .rename({"city":"count"}, axis = 1)
    .sort_values("count", ascending=False)  ## si es nuestro interes organizar los index dados que están desordenados (reset_index)
    .reset_index(drop=True)
)

counts
```

# Volvamos al marco de datos de restarantes

```python
ratings_by_city = restaurants.groupby("City")["Rating"].agg(["mean", "median", "count", "std"]).sort_values("mean", ascending=False)
ratings_by_city
## los primeros son errores dado que son ciudades
```

```python
## Consideremos las ciudad que contenga al menos diez restaurante

ratings_by_city.loc[ratings_by_city["count"] >= 10]
```

```python
## Ahora, dada una combinación de ciudades y rango de precio, ¿donde esta el mejor?

result = restaurants.groupby(["City", "Price Range"])["Rating"].agg(["mean", "median", "count", "std"]).sort_values("mean", ascending=False)
```

```python
result[result["count"]>=10]
```

```python
restaurants.groupby("Price Range")["Rating"].agg(["mean", "median", "count", "std"]).sort_values("mean", ascending=False)
```

```python
## Debemos ordenar los datos para calcular el cambio porcentual
data = data.sort_values(["symbol", "date"])
data
```

```python
#data.drop(columns=["close_pct_change"], axis = 1)
```

```python
## Ahora si la necesidad es calcular el cambio porcentual de una fila respecto a la fila anterior
data["close_pct_change"] = data.groupby("symbol")["close"].pct_change()
data
```

```python
## Verifiquemos si el groupby a quedado correcto y para cualquier empresa debe iniciar en NAN

data[data["symbol"] == "AAPL"]
### data = data['symbol'].value_counts()
```

```python
### Como aplicar transformaciones despues de agrupar

data["close"].transform(["pct_change", "cumsum"]) ## transfor sin agrupar / pct_change = cambio porcentual
```

```python
data.groupby("symbol")["close"].transform("pct_change") ## transfor con agrupar
## obs : Cuando realizas la transform en grupo, no devuelve los simbolos de agrupacion
```
