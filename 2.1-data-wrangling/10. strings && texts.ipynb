## Función final

* Caracteres especiales (@, #, -).

* Espacios múltiples (\s+).

* Números dentro de texto (\d).

* Combinaciones específicas (ej: r'^[A-Z]' para letras mayúsculas al inicio).

* el guión (-) en r'[a-z]' representa un rango de letras (de la "a" a la "z").

```python
Si solo quieres detectar espacios al inicio, usa r'^\s+'.
Si solo quieres detectar espacios al final, usa r'\s+$'.
Encuentra cualquier carácter que no sea una letra, número o espacio en blanco r'[^a-zA-Z0-9\s]'

```

````python
from collections import Counter

def detectar_ruido(df, columna):
    print(f"\n=== Análisis de '{columna}' ===")

    # Espacios al inicio/final (r: al inicio)
    espacios = df[df[columna].str.contains(r'^\s+|\s+$', regex=True, na=False)] ## regex= True(representa expresiones regulares; true indica que cadenas que contengan), False (cadenas que sean exactamente igual)
    print(f"Espacios al inicio/final: {len(espacios)} registros")

    # Caracteres no alfanuméricos
    no_alfanumericos = df[df[columna].str.contains(r'[^a-zA-Z0-9\s]', regex=True, na=False)]
    print(f"Caracteres especiales: {len(no_alfanumericos)} registros")

    # Frecuencia de caracteres
    todos_caracteres = "".join(df[columna].dropna().astype(str))
    contador = Counter(todos_caracteres)
    print("Caracteres más frecuentes:", contador.most_common(5))

    # Ejemplo de registro problemático
    if len(no_alfanumericos) > 0:
        print("\nEjemplo de registro con ruido:")
        print(no_alfanumericos.head(1).values[0])
````

````python
# Aplicar la función a una columna
detectar_ruido(df, "A")
````
