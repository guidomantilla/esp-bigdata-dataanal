```python
forex = forex_data.sort_values("date", ascending = True)
```

```python
forex.loc[:, "EUR":"VEF"].diff() ## podemos diferenciar 1 o n veces diff(n)
```

```python
## Ahora que ocurre si quiero guardar en la base origial solo algunas transformaciones
forex.iloc[:, 1:3] = forex.iloc[:, 1:3].diff()
forex.head()
```

```python
EUR_data = forex[["date", "EUR"]].copy()
EUR_data
```

```python
## Calcular los lag o retardos de una variable.
EUR_data["EUR_lag1"] = EUR_data["EUR"].shift(1) ## shift(3) retardo tres días // shift(-1) adelantarlo un día
EUR_data
```

```python
## Ahora si quiero conocer el porcentaje de cambio o cambio porcentual
EUR_data["EUR_pct_change"] = EUR_data["EUR"].pct_change(1) ## cambio porcentual de dos días
EUR_data
```

```python
## Otra función importante es la suma acumulada
EUR_data.EUR.cumsum()
```

```python
## Producto acumulado
EUR_data.EUR.cumprod()
```

```python
## Ahora que sucede si queremos aplicar multiples funciones
# EUR_data.EUR.transform("shift", 0, 3) ## periodos de retardos
EUR_data.EUR.transform(["shift", "cumsum", "diff"], 0) ## axis = 0 indica que vamos a recorrer las columnas
```

```python
## Cuando se solicita la transformación de más de una funcion se conoce como proceso multiniveles
result1 = forex_data[["EUR", "GBP", "JPY"]].transform(["shift", "cumsum", "diff"], 0)
```

```python
result1
```

```python
result1.columns
```

```python
list(result1.columns)
```

```python
result1.columns = [ x[0]+"_"+x[1] for x in list(result1.columns)]
result1
```

```python
## Formato diccionario para realizar transformaciones a determinadas variables
forex_data.transform({"EUR":"shift", "JPY":"cumsum"})
```

### Definir nuestras propias funcionesde transformación

```python
def add_1(x):
  return x + 1

EUR_data["added_1"] = EUR_data["EUR"].transform(add_1)
EUR_data
```

```python
# User Defined Functions for transformation
def add_n(x, n, m):
    return x+n+m

EUR_data["added_10"] = EUR_data.EUR.transform(add_n, axis=0, n=10, m=100)
EUR_data
```

```python
## Estandarización de datos
realtor_data
```

```python
# (value - mean())/std()
def standardize(x):
    z_score = (x - x.mean())/x.std()
    return z_score
```

```python
standardize(realtor_data.price)
```

```python
realtor_data[["price", "bed", "bath", "house_size"]].transform([standardize])
```

```python
realtor_data[["price", "bed", "bath", "house_size"]].transform([standardize, "cumsum"])
## Ob: Las funciones definidas por el usuario deben ir sin comillas porque se refieren aun objeto
```

# funciones Apply (Aplicar y Mapear funciones)

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("realtor-data.csv", parse_dates=["prev_sold_date"])
data.info()
```

```python
def max_min(x):
  return pd.Series({"max": x.max(), "min": x.min()}) ## creamos un diccionario
```

* La funcón Apply puede comportarse como una agregación o transformación, depende como la utilices.

```python
max_min(data["bed"])
```

```python
## Ahora,como podemos aplicar la función ya sea a filas o columnas
data[["bed", "bath", "acre_lot", "house_size", "price"]].apply(max_min, axis=0) ## axis = 0 significa que voy aplicarlo columna por columna
```

# Función map (aplicar una función para cada celda individual $x_{ij}$)

```python
data.head()
```

```python
## Este tipo de procesos necesitan mucho recurso computaciones pues va uno a uno
## Este tipo de procedimiento es necesario cuando las variables no estás vectorizadas

def count_len(x):
  return len(str(x))

data.map(count_len)
```

# Función Lambda (simplifica la definición de funciones)

```python
data.map(lambda x: len(str(x))) ## Lambda solo acepta funciones de un solo paso
```
